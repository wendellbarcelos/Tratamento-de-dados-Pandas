```python
# Importando a biblioteca
import pandas as pd
```

```python
# Carregando os dados de um arquivo Json, gerou uma tabela fonte com suas chaves e valores
dados_hospedagem = pd.read_json('dados_hospedagem.json')
dados_hospedagem
```

```python
# Ã‰ necessÃ¡rio fazer a normalizaÃ§Ã£o desses dados para que eles gere as colunas de forma correta,
dados_hospedagem = pd.json_normalize(dados_hospedagem['info_moveis'])
dados_hospedagem.head()
```

```python
# Retornando as colunas para comparar qual as colunas que estÃ£o em lista
colunas = list(dados_hospedagem.columns)
colunas
```

```python
# Usando o explode para remover as lista das colunas 
dados_hospedagem = dados_hospedagem.explode(colunas[3:])
```

```python
# Visualizando
dados_hospedagem.head()
```

```python
# NecessÃ¡rio resetar o index
dados_hospedagem.reset_index(inplace=True, drop=True)
```

```python
# Visualizando
dados_hospedagem.head()
```

```python
# Verificando as informaÃ§Ãµes dos dados - todos os dados estÃ£o como object necessÃ¡rio tratar valores
dados_hospedagem.info()
```

```python
# Lista de colunas para alteraÃ§Ã£o de tipo
col_alterar = ['max_hospedes','quantidade_banheiros','quantidade_quartos','quantidade_camas']
col_regex = ['taxa_deposito','taxa_limpeza','preco']
```

```python
# Convertendo os valores numericos com o astype
dados_hospedagem[col_alterar] = dados_hospedagem[col_alterar].astype(int)
dados_hospedagem['avaliacao_geral'] = dados_hospedagem['avaliacao_geral'].astype(float)

# As colunas taxa_deposito, taxa_limpeza, preco tem um digito '$' sendo necessÃ¡rio um regex para a retirada
dados_hospedagem[col_regex] = (dados_hospedagem[col_regex].replace(r'[^\d.]', '', regex=True).astype(float).astype(int))
```

```python
# Visualizando as informaÃ§Ãµes
dados_hospedagem.info()
```

```python
# Visualizando dataframe
dados_hospedagem.head()
```

#### Trabalhando com as colunas de texto

```python
# Colocando os textos em minusculos
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.lower()

# Alterando a coluna com regex para remover todos os caracteres alem do texto, ifen e a apostases
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.replace(r'[^a-zA-Z0-9\-\']', ' ', regex=True)

# Agora esse regex irÃ¡ verificar se hÃ¡ texto antes e depois do ifen senao ele remove
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.replace(r'(?<!\w)-(?!\w)', '', regex=True)
```

```python
# Criar tokenizaÃ§Ã£oa partir do texto com o split
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.split(' ')
```

```python
# Agora o tratamento serÃ¡ na coluna comodidades para gerar a tokenizaÃ§Ã£o nela
dados_hospedagem['comodidades'] = dados_hospedagem['comodidades'].str.replace(r'\{|}|\"','',regex=True)
dados_hospedagem['comodidades'] = dados_hospedagem['comodidades'].str.split(',')
```

```python
# Trabalhando a coluna descricao_vizinhanca
dados_hospedagem['descricao_vizinhanca'] = dados_hospedagem['descricao_vizinhanca'].str.replace(r'[^a-zA-Z0-9\-\']', ' ', regex=True)
dados_hospedagem['descricao_vizinhanca'] = dados_hospedagem['descricao_vizinhanca'].str.replace(r'(?<!\w)-(?!\w)', '', regex=True)
```

```python
# Criar tokenizaÃ§Ã£oa partir do texto com o split
dados_hospedagem['descricao_vizinhanca'] = dados_hospedagem['descricao_vizinhanca'].str.split(' ')
```

```python
# Visualizando dataframe
dados_hospedagem.head()
```

### Trabalhando com os moveis disponiveis

```python
# Carregando os dados
dados_moveis = pd.read_json('moveis_disponiveis.json')
dados_moveis
```

```python
# Precisando usar a coluna de data percebo que nÃ£o esta no formato de data, sendo necessÃ¡rio alterar
dados_moveis.info()
```

```python
# Usando a funÃ§Ã£o do pandas datetime para alterar a coluna
dados_moveis['data'] = pd.to_datetime(dados_moveis['data'])
```

```python
# Verificando alteraÃ§Ã£o na data
dados_moveis.info()
```

```python
# Usando a funÃ§Ã£o strftime para gerar uma valor de data por mes e ano
dados_moveis['data'].dt.strftime('%Y-%m')
```

```python
# Agora Ã© preciso criar uma contagem de vagas por mÃªs, para sabermos as vagas pela demanda de solicitaÃ§Ãµes de aluguel
# Agrupando os valores por mÃªs para criar um subset
subset = dados_moveis.groupby(dados_moveis['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset
```

```python
# Verificado valores nulos
dados_moveis['preco'].isnull().sum()
```

```python
# Tratando valores nulos na coluna preÃ§o 
dados_moveis['preco'] = dados_moveis['preco'].fillna('0.0')
```

```python
# Visualizando
dados_moveis
```

```python
# Usando 
dados_moveis['preco'] = dados_moveis['preco'].replace(r'[^\d.]', '', regex=True).astype(float)
```
